```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a list of numbers to compute the mean and variance of
x = [1,2,4,6,5,4,0,-4,5,-2,6,10,-9,1,3,-6]
n = len(x)

# compute the population mean
popmean = np.mean(x)
popvar = np.var(x)
# compute a sample mean
sample = np.random.choice(x,size=5,replace=True)
sampmean = np.mean(sample)

# print them
```

```python
# compute lots of sample means

# number of experiments to run


nExpers = 10000

# run the experiment!
sampleMeans = np.zeros(nExpers)
sampleVars = np.zeros(nExpers)
for i in range(nExpers):

  # step 1: draw a sample
  sample = np.random.choice(x,size=15,replace=True)

  # step 2: compute its mean
  sampleMeans[i] = np.mean(sample)
  sampleVars[i] = np.var(sample)

samplemeanVar = np.var(sampleMeans)
sampleVarMean = np.mean(sampleVars)

# show the results as a histogram
plt.hist(sampleMeans,bins=40,density=True)
plt.ylabel('Count')
plt.xlabel('Sample mean')
plt.show()
print(f"popvar: {popvar}")
print(f"samplemeanVar: {samplemeanVar*n}")
print(f"sampleVarMean: {sampleVarMean}")
```

```python
plt.hist(sampleVars,bins=40,density=False)
plt.ylabel('Count')
plt.xlabel('Sample mean')
plt.plot([popvar,popvar], [0,600],'r--',label=popvar)
plt.plot([samplemeanVar*n,samplemeanVar*n], [0,600],'g--',label=samplemeanVar*n)
plt.plot([sampleVarMean,sampleVarMean], [0,600],'b--',label=sampleVarMean)
plt.legend()
plt.show()
```

```python

plt.scatter(np.arange(nExpers),sampleVars)
plt.plot([0,nExpers],[popvar,popvar],'r--',label=popvar)
plt.plot([0,nExpers],[samplemeanVar,samplemeanVar],'g--',label=samplemeanVar*n)
plt.plot([0,nExpers],[sampleVarMean,sampleVarMean],'b--',label=sampleVarMean)
plt.legend()
plt.show()
```
